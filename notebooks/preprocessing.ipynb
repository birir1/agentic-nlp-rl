```python
from pathlib import Path
import pandas as pd
from datasets import Dataset, DatasetDict

# Absolute path to MELD
raw_dir = Path("/root/workspace/agentic_nlp_rl/data/raw/meld/data/MELD")
processed_dir = Path("/root/workspace/agentic_nlp_rl/data/processed/meld")

splits = ["train", "validation", "test"]
csv_map = {
    "train": "train_sent_emo.csv",
    "validation": "dev_sent_emo.csv",
    "test": "test_sent_emo.csv"
}

dataset_dict = {}
for split in splits:
    df = pd.read_csv(raw_dir/csv_map[split])
    df = df[['Utterance', 'Emotion']]  # keep only relevant columns
    dataset_dict[split] = Dataset.from_pandas(df)

ds = DatasetDict(dataset_dict)
ds.save_to_disk(processed_dir)
print("MELD preprocessing complete!")
```

```python
from datasets import load_dataset

raw_dir = "/root/workspace/agentic_nlp_rl/data/raw/ultrafeedback/data"
processed_dir = "/root/workspace/agentic_nlp_rl/data/processed/ultrafeedback_prefs"

ds = load_dataset(
    "parquet",
    data_files={
        "train": f"{raw_dir}/train_prefs-00000-of-00001.parquet",
        "test": f"{raw_dir}/test_prefs-00000-of-00001.parquet"
    }
)

ds.save_to_disk(processed_dir)
print("UltraFeedback prefs preprocessing complete!")
```
