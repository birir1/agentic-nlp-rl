```python
## Load MELD
from datasets import load_dataset
from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[0]
MELD_DIR = PROJECT_ROOT / "data" / "raw" / "meld" / "data" / "MELD"

meld = load_dataset(
    "csv",
    data_files={
        "train": str(MELD_DIR / "train_sent_emo.csv"),
        "validation": str(MELD_DIR / "dev_sent_emo.csv"),
        "test": str(MELD_DIR / "test_sent_emo.csv"),
    }
)

meld.save_to_disk(PROJECT_ROOT / "data" / "processed" / "meld")
print(meld)
```
